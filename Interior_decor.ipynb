```python
!pip install diffusers transformers accelerate torch torchvision pillow matplotlib --quiet
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import matplotlib.pyplot as plt

# Load model
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
from google.colab import files
uploaded = files.upload()

# Load image
image_path = list(uploaded.keys())[0]
init_image = Image.open(image_path).convert("RGB")
init_image = init_image.resize((512, 512))

plt.imshow(init_image)
plt.axis("off")
```

```python
prompt = "luxury bedroom, with gold bed vertically allinged, paintings everywhere, a shelve beside the bed"

result = pipe(
    prompt=prompt,
    image=init_image,
    strength=0.7,   # how much change
    guidance_scale=7.5
)

generated_image = result.images[0]

plt.imshow(generated_image)
plt.axis("off")
```
